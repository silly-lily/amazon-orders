```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'data.csv'
df = pd.read_csv(file)
```

## How bad is my Amazon addiction?

### How many orders did I make on Amazon?

```python
total_orders = len(df)
print(total_orders)
```

I made 197 orders on Amazon since I've had an Amazon account.

### How much money did I spend on Amazon orders?

```python
total_spent = 0

for order_cost in df['Item Total']:
    cost = float(order_cost[1:len(order_cost)])
    total_spent+=cost

total_spent = int(np.round(total_spent,0))
print(total_spent)
```

I spent $3241 on Amazon over my lifetime. 

```python
years = 2021-2014
days = years*365 # not inclduing leap years

avg_spent = round(total_spent/days,2)
print(avg_spent)
```

I spent on average $1.27 a day on Amazon purchases

## What was I even buying on Amazon?

```python
orders = []

for order in df["Title"]:
    if type(order) == str:
        orders.append(order)

print(orders)
```

```python
cat = []
cat_num = []

for order in df["Category"]:
    if type(order) == str:
        if order in cat:
            cat_num[cat.index(order)]+=1
        else:
            cat.append(order)
            cat_num.append(1)

biggest_cat = cat[cat_num.index(max(cat_num))]
print(biggest_cat)
```

My most ordered category was pet supplies. During the time I have had Amazon, I have had numerous different types of saltwater fish and three different dogs!

## How did my spending change over the years?


### Orders per year

```python
year = []
counts = []

for order in df["Order Date"]:
    if type(order) == str:
        if '20'+order[6:8] not in year:
            year.append('20'+order[6:8])
            counts.append(1)
        else:
            index = year.index('20'+order[6:8])
            counts[index]+=1

plt.bar(year, counts)
plt.xlabel('Year')
plt.ylabel('Amazon Orders')
plt.title('Amazon Orders by Year')

plt.show()
```
